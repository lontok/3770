```python
import pandas as pd
from matplotlib import pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from IPython.display import display, HTML
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/guinearice.csv', parse_dates=['date'], index_col='date')
df.plot(title='Trend Only', legend=False)

result = seasonal_decompose(df, model='multiplicative', extrapolate_trend='freq')
result.plot()
plt.show()
```
